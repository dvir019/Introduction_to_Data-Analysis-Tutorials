```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Tutorial_6_data.csv')
df.head(20)
```

In this data set there are 11 columns:
* Rank - Ranking of overall sales
* Name - The game's name
* Platform - Platform of the game's release (i.e. PC, PS4, etc.)
* Year - Year of the game's release
* Genre - Genre of the game (i.e. Action, Sports, Role-Playing, etc.)
* Publisher - Publisher of the game
* NA_Sales - Sales in North America (in millions)
* EU_Sales - Sales in Europe (in millions)
* JP_Sales - Sales in Japan (in millions)
* Other_Sales - Sales in the rest of the world (in millions)
* Global Sales - Total worldwide sales

### Who are the 5 most global selling publishers?

```python
top_selling_publishers_df = df_clean.groupby('Publisher', as_index=False)['Global_Sales'].sum() 
# .reset_index() # We can alternatively use the reset_index() method instead of providing the as_index parameter
top_selling_publishers_df
```

```python
top_selling_publishers_df.sort_values(by='Global_Sales', ascending=False, inplace=True)
top_selling_publishers_df
```

```python
top5_selling_publishers_df = top_selling_publishers_df[:5]
top5_selling_publishers_df
```

What if we want to change a column's name?

```python
top5_selling_publishers_df.rename(columns={'Global_Sales': 'Total Sales'}, inplace=True)
top5_selling_publishers_df.head()
```

```python
sns_object = sns.catplot(kind='bar', data=top5_selling_publishers_df, x='Publisher', y='Total Sales')
# sns_object.set_xticklabels(rotation=45) # Rotate the x label ticks text by 90 degrees
# sns_object.fig.suptitle('Top 5 Global Selling Publishers') # set figure title
# sns_object.set(ylabel='Global Sales in Millions') # set y label
# sns_object.fig.set_size_inches(15,10) # set figure size
```

```python
clean_df_copy = df_clean.copy()
clean_df_copy['Year']=clean_df_copy['Year'].astype(int)
global_sales_by_year_df = clean_df_copy.groupby('Year', as_index=False)['Global_Sales'].sum()
sns_global_sales_by_year = sns.catplot(kind='bar', data=global_sales_by_year_df, x='Year', y='Global_Sales')
sns_global_sales_by_year.set_xticklabels(rotation=45) # Rotate the x label ticks text by 90 degrees
sns_global_sales_by_year.fig.suptitle('Global Sales By Year') # set figure title
sns_global_sales_by_year.fig.set_size_inches(20,15) # set figure size
```

```python
game_releases_per_year_df = clean_df_copy.groupby('Year', as_index=False)['Global_Sales'].count()
sns_game_releases_by_year = sns.catplot(kind='bar', data=game_releases_per_year_df, x='Year', y='Global_Sales')
sns_game_releases_by_year.set_xticklabels(rotation=45) # Rotate the x label ticks text by 90 degrees
sns_game_releases_by_year.fig.suptitle('Game Releases Pe Year') # set figure title
sns_game_releases_by_year.set(ylabel='Game Releases') # set y label
sns_game_releases_by_year.fig.set_size_inches(20,15) # set figure size
```
